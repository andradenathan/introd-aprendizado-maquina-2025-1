```python
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_csv('data.csv')
```

Questão 2b)

É possível notar aqui que a sétima observação é um outlier dado que o valor "450" é bem distante da média de duração, puxando para um valor muito alto.

```python
zscore(df['Duration'])
```

Aqui, faremos a substituição da duração que está com um valor extremo para a média, de forma que não altera o nosso resultado pois não altera o resultado da nossa análise.

```python
mean_durations = df["Duration"].mean()

df.loc[7, "Duration"] = int(mean_durations)
```

Queremos preencher as calorias, para isto, utilizaremos a média em lugares que os valores são vazio.

```python
median_calories = df["Calories"].median()

df["Calories"] = df["Calories"].fillna(median_calories)
```

Assim como as datas estão mal formatadas e principalmente com valores numéricos (erratas), queremos fazer esse tratamento transformando para data. O que for transformado e falhar, isto é, um dado que não é uma data, eliminaremos a linha.

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')

df.dropna(subset=['Date'], inplace = True)
```

Eliminaremos dados duplicados pois não será útil para o nosso dataset final manter as mesmas instâncias.

```python
df.drop_duplicates(inplace = True)

df
```

Questão 3)

Sabemos que ao calcular o z-score, temos que se $z > 3$ ou $z < -3$, então, os valores estão a mais de 3 desvios padrões da média (para cima ou para baixo), pois a fórmula:

$z = \frac{x - \mu}{\sigma}$

Aqui na coluna "Pulse", a linha 23 tem um z-score maior do que 3, o que indica que é um bom indício para considerar outlier.

```python
zscore(df['Pulse'])
```

Questão 3)

Aqui na coluna "Maxpulse", a linha 3 tem um z-score maior do que 3, o que indica que é um bom indício para considerar outlier.

```python
zscore(df['Maxpulse'])
```

Questão 3)

Não há indício de outlier em calorias por conta dos critérios estabelecidos anteriormente.

```python
zscore(df['Calories'])
```

```python
z_scores = df[['Pulse', 'Maxpulse', 'Calories']].apply(zscore)

combinacoes = (z_scores < 3).all(axis=1) & (z_scores > -3).all(axis=1)

df = df[combinacoes]

df
```
