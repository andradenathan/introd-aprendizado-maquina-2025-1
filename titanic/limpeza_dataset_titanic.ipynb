```python
import pandas as pd

titanic_train_df = pd.read_csv('train.csv')
```

Nosso primeiro passo na limpeza dos dados para tratamento do conjunto de treinamento, será fazer o levantamento estatístico e informativo dos dados.

```python
titanic_train_df.describe()
```

```python
titanic_train_df.isnull().sum()
```

Para tratar o caso das idades nulas, iremos criar um novo atributo chamado "Age_1", onde iremos calcular a mediana pois é uma boa aproximação de modo que não interferirá em nossa análise.

Queremos analisar a idade das pessoas que puderam pagar por tickets diferentes e fazer separação por gênero de pessoa.

```python
titanic_train_df.Age.describe()
```

```python
titanic_train_df.groupby(['Pclass', 'Fare', 'Sex'])['Age'].median()
```

```python
titanic_train_df["Age_1"] = titanic_train_df.Age.fillna(titanic_train_df.groupby(['Pclass', 'Fare', 'Sex'])['Age'].transform('median'))
```

```python
titanic_train_df["Age_2"] = titanic_train_df['Age_1'].fillna(titanic_train_df['Age'].median())
```

```python
titanic_train_df.head(3)
```

```python
titanic_train_df.isnull().sum()
```

Levando em consideração que queremos fazer uma análise na distribuição de pessoas que puderam pagar por tickets diferentes, e se sobreviveram ao Titanic de fato, não estamos interessados nas features de Cabine e nem de porto de embarque.

Manteremos o **passengerId** por fins de análise para identificar os dados como um "rótulo" caso queiramos armazenar esses dados em um banco posteriormente.

Além disso, não precisamos do número do bilhete, visto que é suficiente saber a sua classe.
